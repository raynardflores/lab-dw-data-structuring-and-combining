## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd

# URLS
file1_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
file2_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
file3_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# IMPORTING
data1 = pd.read_csv(file1_url)
data2 = pd.read_csv(file2_url)
data3 = pd.read_csv(file3_url)

# COMBINED DATA WITH CONCAT
combined_data = pd.concat([data1, data2, data3], ignore_index=True)

# Changing the null values with the median

combined_data['Income'].fillna(combined_data['Income'].median(), inplace=True)
combined_data['Monthly Premium Auto'].fillna(combined_data['Monthly Premium Auto'].median(), inplace=True)
combined_data['Total Claim Amount'].fillna(combined_data['Total Claim Amount'].median(), inplace=True)

# Changing null values of categorical columns with the mode

combined_data['ST'].fillna(combined_data['ST'].mode()[0], inplace=True)
combined_data['GENDER'].fillna(combined_data['GENDER'].mode()[0], inplace=True)
combined_data['Number of Open Complaints'].fillna(combined_data['Number of Open Complaints'].mode()[0], inplace=True)
combined_data['State'].fillna(combined_data['State'].mode()[0], inplace=True)
combined_data['Gender'].fillna(combined_data['Gender'].mode()[0], inplace=True)

# Replacing the name of the values

combined_data['ST'].replace({'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'}, inplace=True)

combined_data.rename(columns={'ST': 'State_2'}, inplace=True)

# Cleaning'
combined_data['Customer Lifetime Value'] = combined_data['Customer Lifetime Value'].str.replace('%', '')

# Changing the type to float
combined_data['Customer Lifetime Value'] = combined_data['Customer Lifetime Value'].astype(float)

# Replacing null values with the mean
mean_lifetime_value = combined_data['Customer Lifetime Value'].mean()
combined_data['Customer Lifetime Value'].fillna(mean_lifetime_value, inplace=True)

# Rounding the values two decimals
combined_data['Customer Lifetime Value'] = combined_data['Customer Lifetime Value'].round(2)

# Changing the type
combined_data['Income'] = combined_data['Income'].astype('Int64')

mean_income = combined_data['Income'].mean()

combined_data['Income'].fillna(int(mean_income), inplace=True)

combined_data
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Challenge 2

import pandas as pd

# Load the dataset

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
data = pd.read_csv(url)

# Display the first few rows of the dataset to understand its structure
data.head()

# Deleting a row 

data.drop('unnamed:_0', axis=1, inplace=True)

# Total revenue for each sales channel

sales_channel_revenue = data.pivot_table(index='sales_channel', values='total_claim_amount', aggfunc='sum')
sales_channel_revenue['total_revenue'] = round(sales_channel_revenue['total_claim_amount'], 2)
print("\nTotal revenue for each sales channel:")
print(sales_channel_revenue[['total_revenue']])

# Average customer lifetime value per gender and education level

avg_clv_gender_education = data.pivot_table(index=['gender', 'education'], values='customer_lifetime_value', aggfunc='mean')
print("\nAverage customer lifetime value per gender and education level:")
print(avg_clv_gender_education)
```
